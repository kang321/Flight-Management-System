```python
import pandas as pd 
import numpy as np 
```

```python
def cache(source1, destination1,date1):
    df = pd.read_csv('kayak1_data.csv')
    df1 = df[(df["Source"] == source1) & (df["Destination"]==destination1) & (df["Date"] == date1)]
    df1.reset_index(inplace=True)
    return df1
    
```

```python
source = "AMD"
destination ="YYZ"
date="2022/05/06"

df1 = cache(source, destination, date)
df1
```
